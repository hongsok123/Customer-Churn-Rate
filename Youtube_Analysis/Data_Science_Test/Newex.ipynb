```python
df.TotalSales.plot(kind = "hist", grid = True)
plt.title("The total Sale")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.legend(loc =1)
plt.show()
```

```python
import matplotlib.patches as mpatches
df.TotalSales.plot(kind = "box")
red_path = mpatches.Patch(color = 'blue', label = "TotalSales")
plt.legend(handles = [red_path], loc = 1)
plt.show()
```

```python
sns.boxplot(y= df.TotalSales)
plt.show()
```

```python
df.plot( x = 'TotalSales' , y= 'OperatingCost', kind = "scatter")
plt.show()
```

```python
sns.countplot( data = df, x = df["StoreType"])
```

```python
sns.scatterplot(x = df.TotalSales , y = df.OperatingCost, color= "red", hue = df.Location)
```
