6. ### Create a new variable called “customer value segment

```python
Customer_Segment = []
for row in df['Customer_Value']:
    if row < 10000 :    Customer_Segment.append('Low Value Segment')
    elif row > 10000 and row < 250000 :   Customer_Segment.append('Medium Value Segment')
    elif row > 10000 : Customer_Segment.append('High Value Segment')
        
    else: Customer_Segment.append('Not_Rated')

df['Customer_Segment'] = Customer_Segment
```

```python
df[['Customer_Value', 'Customer_Segment']]
```
